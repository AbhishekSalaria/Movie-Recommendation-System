## CONTENT BASED RECOMMENDATION (TMDB 5000 Movie Dataset : kaggle)


```python
import numpy as np
import pandas as pd
```

### IMPORTING DATA

```python
movies = pd.read_csv("tmdb_5000_movies.csv")
credits = pd.read_csv("tmdb_5000_credits.csv")
```

### PREPROCESSING DATA

```python
def preprocess(data):
    result = []
    for dictionary in eval(data):
        result.append(dictionary['name'])
    return result
```

```python
movies['genres'] = movies['genres'].apply(preprocess)
```

```python
movies['production_companies'] = movies['production_companies'].apply(preprocess)
```

```python
movies['cast'][0]
```

```python
movies['cast'] = movies['cast'].apply(preprocess)
```

```python
movies['cast'][0]
```

```python
movies['crew'][0]
```

```python
def processCrew(data):
    result = []
    for dictionary in eval(data):
        if dictionary['job'] in ["Producer", "Director"]:
            result.append(dictionary["name"])
    return result
```

```python
movies['crew'] = movies['crew'].apply(processCrew)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x: x.split())
```

```python
movies['overview'][0]
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['overview'] = movies['overview'].apply(lambda x : [i.replace(" ","") for i in x])
movies['production_companies'] = movies['production_companies'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['alltags'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['production_companies'] + movies['cast'] + movies['crew']
```

```python
movies['alltags'][0]
```

```python
movies['alltags'] = movies['alltags'].apply(lambda x: " ".join(x))
movies['alltags'] = movies['alltags'].apply(lambda x: x.lower())
```

```python
preprocessed_data = movies[['id','original_title','alltags']]
```

```python
preprocessed_data.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
stemmer = PorterStemmer()

def stemming(data):
    result = []
    for word in data.split():
        result.append(stemmer.stem(word))
    return " ".join(result)
```

```python
preprocessed_data['alltags'] = preprocessed_data['alltags'].apply(stemming)
```

```python
preprocessed_data['alltags'][0]
```

### TRANSFORMING TAGS INTO VECTORS (BAG OF WORDS)

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
vectorizer = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = vectorizer.fit_transform(preprocessed_data['alltags']).toarray()
```

```python
vectors.shape
```

```python
vectorizer.get_feature_names()
```

### FINDING COSINE SIMILARITY OF EACH MOVIE WITH EVERY OTHER MOVIE

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity
```

### RECOMMENDING TOP 5 MOVIES HAVING HIGHEST COSINE SIMILARITY TO A GIVEN MOVIE

```python
def recommendation(movie):
    index = preprocessed_data[preprocessed_data["original_title"]  == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse = True, key = lambda x: x[1])[1:6]
    for data in distances:
        print(preprocessed_data.iloc[data[0]].original_title)
```

### PREDICTIONS

```python
recommendation("Avatar")
```

```python
recommendation("Terminator 2: Judgment Day")
```

```python
recommendation("Pirates of the Caribbean: At World's End")
```
